```python
import pandas as pd
import numpy as np

df = pd.read_csv('../Data/Border_Crossing_Entry_Data.csv')
```

### Data Cleaning

```python
# performing null check
df.isnull().sum()
```

```python
# looking for duplicates
df.duplicated().any()
```

```python

# displaying duplicates
duplicates = df[df.duplicated(keep=False)]
grouped = duplicates.groupby(list(df.columns))
idx = [x.index for _, x in grouped if len(x) > 1]
duplicates.loc[pd.Index(np.concatenate(idx))]

```

```python
# removing duplicates
df = df.drop_duplicates()
# checking for duplicates
df.duplicated().any()
```

```python
# saving cleaned dataframe
df.to_csv('../Data/Cleaned_Border_Crossing_Entry_Data.csv')
```
